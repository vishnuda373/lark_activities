```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Dataset_w6.xlsx")
data
```

# Find out how many countries are there in the dataset.

```python
country = data.country.unique()
# print(country)
size = country.size
print(size)
```

# Examine the data by using visualizations.

```python
largest = data.nlargest(10,["Population in thousands (2017)"])
# print(largest)
# plt.figure(figsize=(6, 7))
largest.plot(x = "country", y = "Population in thousands (2017)", kind = "bar")
plt.xticks(rotation = 60 )
plt.show()
```

# Find out how many countries with respect to the region are available in the dataset

```python
data.groupby('Region').size()
```

# On average which region has the highest                                                           

# a. GDP per capita (current US$)

```python
highest_gdp = data.nlargest(1,["GDP per capita (current US$)"])
highest_gdp
```

# b. International trade: Imports (million US$)

```python
highest_trade = data.nlargest(1,["International trade: Imports (million US$)"])
highest_trade
```

# Find out which region has more consistent with respect to GDP per capita (current US$)

```python
mean_data = data.groupby("Region")["GDP per capita (current US$)"].mean()
# mean_data
consistent_region = mean_data.nlargest(1)
print(consistent_region)
```
