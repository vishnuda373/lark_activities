```python
import pandas as pd
import numpy as np

data = pd.read_csv("diabetes_model.csv")
min_value = data["BMI"].min(axis = 0)
print("Minimum Value of BMI :",min_value)

mean_value = data["BMI"].mean()
print("\n")
print("Mean Value of BMI :",mean_value)

index = data[data["BMI"] == min_value].index.values
print("\n")
print("Index Having Minimum Values :",index)

updated_data = data.replace([min_value],mean_value) 
# print("\n")
# print("Updated Data Set \n")
# print(updated_data)

# showing the values are updated
show = updated_data["BMI"].values[[index]]
print("\n")
print("After Update :",show)
```

### 5.

**-** Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

**-** Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

**-** Combine the arrays first and second vertically, the final array size should be 10 x 5

```python
import numpy as np

array_1 = np.zeros((5,5))
print("Array of 0s")
print(array_1)

array_2 = np.ones((5,5))
print("\nArray of 1s")
print(array_2)

result_array = np.concatenate((array_1,array_2),axis = 0)
print("\n")
print("Result")
print(result_array)
```
